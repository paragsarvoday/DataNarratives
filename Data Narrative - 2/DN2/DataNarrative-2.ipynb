```python
#Cleaning the columns
college.drop(college[college["Out-of-state tuition"] == "*"].index, inplace=True)
college.drop(college[college["Room and board costs"] == "*"].index, inplace=True)
college.drop(college[college["Avg. Combined SAT score"] == "*"].index, inplace=True)
college.drop(college[college["Number of applicants accepted"] == "*"].index, inplace=True)
college.drop(college[college["Number of applications received"] == "*"].index, inplace=True)
college.drop(college[college["Instructional expenditure per student"] == "*"].index, inplace=True)
college.drop(college[college["Graduation rate"] == "*"].index, inplace=True)
college.drop(college[college["Pct. of faculty with Ph.D.'s"] == "*"].index, inplace=True)
college.drop(college[college["Student/faculty ratio"] == "*"].index, inplace=True)

```

```python

#selected_columns = ["Avg. Combined SAT score", "Out-of-state tuition","State (postal code)"]
lst = ["Room and board costs", "Number of applications received", "Number of applicants accepted", "Avg. Combined SAT score", "Instructional expenditure per student", "Graduation rate", "Pct. of faculty with Ph.D.'s","Student/faculty ratio" ]
college[lst] = college[lst].apply(pd.to_numeric)

acceptance_rate = ((college["Number of applicants accepted"]) / (college["Number of applications received"])) * 100
#acceptance_rate_df = pd.DataFrame(acceptance_rate)

selected_columns = ["Avg. Combined SAT score", "Out-of-state tuition"]

df1 = college[selected_columns].copy()
df1["Out-of-state tuition"] = pd.to_numeric(df1["Out-of-state tuition"])
df1["Acceptance rate"] = acceptance_rate

df1["Out-of-state tuition"] = college["Room and board costs"] + df1["Out-of-state tuition"]
# df1 = df1.reset_index()
df1.columns = ["Avg. SAT score", "Total expenses", "Acceptance rate"]
print(df1)
# print(type(df1))
```

#### Trying to find any co-relation between instructional expenditure per student and total fees ####

```python
df1["Instructional expenditure per student"] = college["Instructional expenditure per student"]
import sklearn
correlation = df1["Total expenses"].corr(df1["Instructional expenditure per student"])
print(f"The correlation between the parameters is {correlation}")
```

```python
# plotting the data
plt.scatter(df1["Instructional expenditure per student"], df1["Total expenses"])
 
# This will fit the best line into the graph
plt.plot(np.unique(df1["Instructional expenditure per student"]), np.poly1d(np.polyfit(df1["Instructional expenditure per student"], df1["Total expenses"], 1))
         (np.unique(df1["Instructional expenditure per student"])), color='red')

plt.title('Correlation between total fees paid by a student and instructional cost')
plt.xlabel('Instructional expenditure per student')
plt.ylabel('Total fees paid by the student')
```

There exists strong positive corrrelation (âˆ¼0.64) between the fees paid and the value of services received from the universities. This shows that atleast in 1994, the universities in US were reasonably priced.

#### Exploring relationship between graduation rate and acceptance rate ####

```python
df1["Graduation rate"] = college["Graduation rate"]
df1.drop(df1[df1["Graduation rate"] > 100].index, inplace=True)
print(df1)

correlation = df1["Graduation rate"].corr(df1["Acceptance rate"])
print(f"The correlation between the parameters is {correlation}")
```

```python
a = df1.loc[df1['Graduation rate'] > 100]
print(a)
```

```python
# plotting the data
plt.scatter(df1["Acceptance rate"], df1["Graduation rate"])
 
# This will fit the best line into the graph
plt.plot(np.unique(df1["Acceptance rate"]), np.poly1d(np.polyfit(df1["Acceptance rate"], df1["Graduation rate"], 1))
         (np.unique(df1["Acceptance rate"])), color='red')

plt.title('Correlation between acceptance rate and graduation rate')
plt.xlabel('Acceptance rate')
plt.ylabel('Graduation rate')
```

The two chosen parameters are negatively corelated (âˆ¼(-0.22)) which means lower the acceptance rate, higher the graduation rate. Or pick wisely, reap profusely.

The universities which have low acceptance rate or which are more selective in their admission process are more successful in picking out the right talent. These students continue to perform well in academics which at the end results in high graduation rate.
