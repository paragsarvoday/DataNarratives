Trying to apply K-means clustering on the basis of round numbers

```python
from sklearn.cluster import KMeans

round_data = aus_men['Round']

round_data = np.array(round_data).reshape(-1, 1)

kmeans = KMeans(n_clusters=7, random_state=0).fit(round_data)

labels = kmeans.labels_

plt.scatter(round_data, np.zeros_like(round_data), c=labels)
plt.xlabel('Round')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Coverting all the entries into int data type
aus_men = aus_men.astype({"Round":"int","TPW.1":"int", "TPW.2":"int"})

# Select the features relevant for clustering
rounds = aus_men['Round']
total_points_won = aus_men['TPW.1'] + aus_men['TPW.2']

# Combine the features into a single dataframe
X = pd.concat([rounds, total_points_won], axis=1)
#Converting the column names to str data type
X.columns = X.columns.astype(str)

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Apply k-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X_scaled)

# Visualize the clusters
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=pred_y)
plt.title('Clustering Matches Based on Round')
plt.xlabel('Round')
plt.ylabel('Total Points Won')
plt.show()
```

Trying the same with number of aces

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Coverting all the entries into int data type
aus_men = aus_men.astype({"Round":"int","TPW.1":"int", "TPW.2":"int"})

# Select the features relevant for clustering
rounds = aus_men['Round']
total_points_won = aus_men['ACE.1'] + aus_men['ACE.2']

# Combine the features into a single dataframe
X = pd.concat([rounds, total_points_won], axis=1)
#Converting the column names to str data type
X.columns = X.columns.astype(str)

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Apply k-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X_scaled)

# Visualize the clusters
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=pred_y)
plt.title('Clustering Matches Based on Round')
plt.xlabel('Round')
plt.ylabel('Total number of aces')
plt.show()
```

Seems like the variation in the total number of aces in the the initial rounds is quite high. This might be because players of various different strengths are paiired against each other in initial rounds. But later on as players of similar skill level are matched, the variation becomes close to zero. This shows that from round 3 onwards the difficulty of matches significantly increases as aces are easy points.

```python
# import pandas as pd
# import seaborn as sns
from scipy.stats import pearsonr

# Load the tennis dataset
tennis_df = us_men

# Group the dataset by match and calculate the total break points created and won by each player
grouped_df = tennis_df.groupby(['Player1', 'Player2', 'Round']).agg({
    'BPC.1': 'sum', 'BPW.1': 'sum', 'BPC.2': 'sum', 'BPW.2': 'sum'
}).reset_index()

# Add a new column indicating the match winner
grouped_df['Winner'] = tennis_df.loc[tennis_df['Result'] == 1, 'Player1'].tolist() + \
                       tennis_df.loc[tennis_df['Result'] == 0, 'Player2'].tolist()

# Calculate the total number of break points won by the winner and loser for each match
grouped_df['Winner_BP'] = grouped_df.apply(lambda row: row['BPW.1'] if row['Winner'] == row['Player1'] else row['BPW.2'], axis=1)
grouped_df['Loser_BP'] = grouped_df.apply(lambda row: row['BPW.2'] if row['Winner'] == row['Player1'] else row['BPW.1'], axis=1)

# Plot a scatter plot of the total number of break points won by the winner and loser
sns.scatterplot(data=grouped_df, x='Winner_BP', y='Loser_BP')

# Calculate the Pearson correlation coefficient between the total number of break points won by the winner and loser
corr_coef, p_value = pearsonr(grouped_df['Winner_BP'], grouped_df['Loser_BP'])
print(f"Pearson correlation coefficient: {corr_coef}")
```
